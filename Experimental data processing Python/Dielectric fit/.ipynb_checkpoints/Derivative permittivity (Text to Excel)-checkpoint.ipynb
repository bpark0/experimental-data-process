```python
# This program calculates parameters using dielectric constants over all measured temperature
# from a text file from dielectric spectrometer
# Calculation methods are followed by NIST Derivative File

# Input file MUST have at least four columns in the following order (Frequency, Temperature, Epsilon', Epsilon'')

# This program asks to open a .txt file and generates a .xlsx file.
# you can ignore a RuntimeWarning (it comes with double derivative modulus (dLnLoss))
#/*------------------------------------------------------------------------------*
# * File Name:	NIST Derivative File											*
# * Creation: 	3-29-2010														*
# * Purpose: OriginC Source C file												*
# * 																				*
# * Modification Log:	6-4-2014, J. L. Schaefer														*
# *------------------------------------------------------------------------------*/

# Translated from Origin C to Python: Bumjun Park #
# Creation: 5/31/2020
# The temperature is assumed to be in Celcius
# Ionic conductivity extraction and plot added

# if the program does not work, install packages via terminal
# (e.g.) pip install openpyxl

import numpy as np
import pandas as pd
import openpyxl
from openpyxl.utils.dataframe import dataframe_to_rows
from scipy.signal import savgol_filter
import matplotlib.pyplot as plt
import os
%matplotlib inline

######################################################
################ EDIT FILE NAME HERE #################
######################################################

# Read a text file using the file name
filename = '2-4C18SO2CH3_1107.TXT'

######################################################
######################################################
######################################################

# open the text file
with open(filename, mode='rb') as text:
    contents = text.read()

# convert binary text to string
string = ''
for i in range(len(contents)):
    string += chr(contents[i])

# save the string as lines
lines = string.split('\n')

# get columns
columns = lines[2].split('\t')

# get numbers and save as a table
data_numbers = []
for i in range(len(lines[3:])):
    line_list_str = lines[i+3].split('\t')
    line_list = []
    for j in range(len(line_list_str)):
        line_list.append(float(line_list_str[j]))
    data_numbers.append(line_list)

data_numbers = list(np.array(data_numbers).T)
data = pd.DataFrame()
for i in range(4):
    data[columns[i]] = data_numbers[i]

### separate data by temperature set ###
# lists save each parameter
frequency_list = []
temperature_list = []
eps_real_list = []
eps_imag_list = []

freq_sub = []
temp_sub = []
eps_real_sub = []
eps_imag_sub = []
for i in range(len(data)):
    if len(temp_sub) == 0:
        freq_sub.append(data.iloc[i, 0])
        temp_sub.append(data.iloc[i, 1])
        eps_real_sub.append(data.iloc[i, 2])
        eps_imag_sub.append(data.iloc[i, 3])
    elif data.iloc[i, 1] != data.iloc[i-1, 1]:
        frequency_list.append(freq_sub)
        temperature_list.append(temp_sub)
        eps_real_list.append(eps_real_sub)
        eps_imag_list.append(eps_imag_sub)
        freq_sub = []
        temp_sub = []
        eps_real_sub = []
        eps_imag_sub = []
        freq_sub.append(data.iloc[i, 0])
        temp_sub.append(data.iloc[i, 1])
        eps_real_sub.append(data.iloc[i, 2])
        eps_imag_sub.append(data.iloc[i, 3])
    else:
        freq_sub.append(data.iloc[i, 0])
        temp_sub.append(data.iloc[i, 1])
        eps_real_sub.append(data.iloc[i, 2])
        eps_imag_sub.append(data.iloc[i, 3])
    if i == len(data)-1:
        frequency_list.append(freq_sub)
        temperature_list.append(temp_sub)
        eps_real_list.append(eps_real_sub)
        eps_imag_list.append(eps_imag_sub)

####################################################

### turn lists into numpy arrays ###
frequency_arr = np.array(frequency_list)
temperature_arr = np.array(temperature_list)
eps_real_arr = np.array(eps_real_list)
eps_imag_arr = np.array(eps_imag_list)


# Enable this if the temperature is in Kelvin
"""### convert temperature unit
is_kelvin = input('Are temperatures in Celcius? (y/n)\ny - Celcius\nn - Kelvin\n')
if is_kelvin[0].lower() == 'n' or is_kelvin[0].lower() == 'k':
    temperature_arr -= 273.15"""

# get a temperature list of each dataset in integer
def get_temp_list(temp):
    temp_list = []
    for item in temp:
        if len(temp_list) == 0:
            temp_list.append(item)
        if temp_list[-1] != item:
            temp_list.append(item)
    return temp_list

# get sheet names in temperature, including digit
def gen_sheet_name(temp):
    temp_int = get_temp_list(temp)
    temp_str = []
    num_seg = 1
    for item in temp_int:
        count = 1
        temp_str.append(str(item)+'C')
        for i in range(len(temp_str)-1):
            if item == float(temp_str[i].split(" ")[0][:-1]):
                count += 1
        if count > 1:
            temp_str[-1] = temp_str[-1]+" (%d)" % count
            num_seg = count
    return temp_str, num_seg

# get sheet names in temperature, rounded
def gen_sheet_name_round(temp):
    temp_int = get_temp_list(temp)
    temp_str = []
    num_seg = 1
    for item in temp_int:
        count = 1
        temp_str.append(str(round(item))+'C')
        for i in range(len(temp_str)-1):
            if item == float(temp_str[i].split(" ")[0][:-1]):
                count += 1
        if count > 1:
            temp_str[-1] = temp_str[-1]+" (%d)" % count
            num_seg = count
    return temp_str, num_seg

# get a list of sheet names in temperature
sheet_name_list, num_segment = gen_sheet_name_round(data.iloc[:, 1])
# if up and down print(sheet_name_list[len(sheet_name_list)//2])
print("Number of segments:",num_segment)
log_freq = np.log10(frequency_arr)
const = eps_real_arr
loss = eps_imag_arr
length_array = loss.shape[1]
if loss.shape[1]%2 == 0:
    length_array -= 1

###########deriv loss calculation #############
"""
Calculate derivative loss
"""
deriv_loss = savgol_filter(const, window_length=5, polyorder=2, deriv=1,axis=1) # derivative and smooth
fDiff = log_freq[0,0]-log_freq[0,1] # single value since the gap of log10 freq is uniform by the measurement setting
SGval = np.log(10**fDiff) # d ln f
deriv_loss *= ((np.pi/2)/SGval) # pi/2 * d(eps_real)/d(ln f)

################################################

########## calculate other parameters ##########
tanDelta = loss/const
conduct = 2*np.pi*8.85e-14*loss*frequency_arr
iConduct = 2*np.pi*8.85e-14*const*frequency_arr
storMod = const/(const**2+loss**2)
lossMod = loss/(const**2+loss**2)
################################################

########## calculate derivative Ln Loss #########
"""
Calculate derivative Ln Loss
"""

vLnFreq = np.log(log_freq)
vLoss = loss.copy()
dLnLoss = -np.gradient(vLoss,vLnFreq[0],axis=1)

###########################################################

print("The RuntimeWarning can be ignored without problems.")

# angular frequency
w = 2*np.pi*frequency_arr

# Summary column
numTemps = len(sheet_name_list)
comment1 = "Processed {} Temperatures".format(numTemps)
comment2 = "From File: {}".format(filename)

# make a new xlsx file named same as the text file
wb = openpyxl.Workbook()
dataframe_total = []
df_cond = []
### make dataframe list to convert to each datasheet ###
for i in range(numTemps):
    # save a full data
    cols = {}
    cols["LogFrequency"] = ["Log Frequency","log\-(10) Hz",""]+list(log_freq[i])
    cols["Constant"] = ["Dielectric Constant (\g(e)') {}.\+(o)C".format(sheet_name_list[i][:-1]),"",""]+list(const[i])
    cols["Loss"] = ["Dielectric Loss (\g(e)'') {}.\+(o)C".format(sheet_name_list[i][:-1]),"",""]+list(loss[i])
    cols["DerivLoss"] = ["Derivative Loss (\g(e)''\-(D)) {}.\+(o)C".format(sheet_name_list[i][:-1]),"",""] + list(deriv_loss[i])
    cols["TanD"] = ["Tan-Delta {}.\+(o)C".format(sheet_name_list[i][:-1]),"",""]+list(tanDelta[i])
    cols["Conduct"] = ["Conductivity (\g(s)') {}.\+(o)C".format(sheet_name_list[i][:-1]), "S/cm",""] + list(conduct[i])
    cols["Iconduct"] = ["Conductivity (\g(s)'') {}.\+(o)C".format(sheet_name_list[i][:-1]),"",""] + list(iConduct[i])
    cols["Modulus'"] = ["Storage Modulus {}.\+(o)C".format(sheet_name_list[i][:-1]),"",""] + list(storMod[i])
    cols["Modulus''"] = ["Loss Modulus {}.\+(o)C".format(sheet_name_list[i][:-1]),"",""] + list(lossMod[i])
    cols["DoubleLnLoss"] = ["Ln(\g(e)'')/Ln(\i(f)) {}.\+(o)C".format(sheet_name_list[i][:-1]),"",""] + list(dLnLoss[i])
    cols["Frequency"] =  ["Frequency (Hz) {}.\+(o)C".format(sheet_name_list[i][:-1]),"",""] + list(frequency_arr[i])
    cols["wFrequency"] = ["Angular Frequency {}.\+(o)C".format(sheet_name_list[i][:-1]),"",""] + list(w[i])
    cols["Summary"] = ["Import Comments","","",comment1, comment2] + [""]*(len(list(log_freq[i]))-2)
    df = pd.DataFrame(cols)
    ws = wb.create_sheet()
    ws.title = sheet_name_list[i]
    for r in dataframe_to_rows(df, index=False, header=True):
        ws.append(r)
    
    # get ionic conductivity plateau for each temperature
    dfAdd = df.copy().iloc[3:]
    dfAdd['Temperature'] = sheet_name_list[i]
    dfAdd['TempNum [oC]'] = float(sheet_name_list[i].split()[0][:-1])
    dfAdd['ArrTemp [1000/T (K)]'] = 1000/(273.15+dfAdd['TempNum [oC]'])
    dataframe_total.append(dfAdd)
    df_cond.append(dfAdd[dfAdd['TanD'] == dfAdd['TanD'].max()])
    

df_conductivity = pd.concat(df_cond)
df_conductivity = df_conductivity[['Temperature','ArrTemp [1000/T (K)]','Conduct','Frequency']]
df_conductivity.rename(columns={'Conduct': 'Conductivity [S/cm]'},inplace=True)
df_conductivity.reset_index(inplace=True,drop=True)

# remove default sheet and save the result file as .xlsx in the same directory
wb.remove(wb["Sheet"])

# make a new directory named as the input filename and save all results in the directory
output_dir = os.path.join(os.path.split(filename)[0],os.path.split(filename)[1][:-4])
try:
    os.mkdir(output_dir)
    print("A new directory is created {}".format(output_dir))
except FileExistsError:
    print("The directory already exists.")
output_name = os.path.join(output_dir,os.path.split(filename)[1][:-4])

wb.save(f"{output_name} all data.xlsx")
df_conductivity.to_csv(f"{output_name} ionic conductivity vs T.csv",index=False)

# Data plot to check the data quality

# set default font style and size globally in matplotlib
plt.rcParams['font.size'] = 15
plt.rcParams['font.family'] = 'Times New Roman'

# edit figure size if too small or too big
figure_size = 5

# x and y data for 1000/T plot
fig_x1=df_conductivity['ArrTemp [1000/T (K)]'].values
fig_y1=df_conductivity['Conductivity [S/cm]'].values
fig_x3 = df_conductivity['Frequency']

# x and y data list for conductivity vs freq. plot
fig_x2=frequency_arr
fig_y2=conduct

nseg=num_segment

# apply color maps
n = len(fig_x1)//nseg+1
colors = plt.cm.coolwarm(np.concatenate((np.linspace(0,1,n),np.linspace(1,0,n)))) # Blues

# reverse color list if the first segment is decreasing temperature
if float(df_conductivity["Temperature"][0].rstrip("C")) > float(df_conductivity["Temperature"][1].rstrip("C")):
    colors = plt.cm.coolwarm(np.concatenate((np.linspace(1,0,n),np.linspace(0,1,n))))

def plot_conductivities(x1,y1,x2,y2,x3,nseg):
    fig, axes = plt.subplots(nrows=nseg,ncols=2,figsize=(figure_size*2,figure_size*nseg))
    if nseg == 1:
        axes[0].plot(x1,y1,'ko')
        axes[0].set_title("Segment 1",fontsize=18)
        axes[0].set_xlabel("1000/T $\mathregular{(K^{-1})}$",fontsize=17)
        axes[0].set_ylabel("Ionic conductivity (S/cm)",fontsize=17)
        axes[0].set_yscale("log")
        #axes[0].set_ylim(1e-8,1e-2)
        axes[0].grid(color='grey',ls=':',lw=1)
        for j in range(len(x2)):
            axes[1].plot(x2[j],y2[j],c=colors[j])
            axes[1].plot(x3[j],y1[j],marker='x',c='k')
        axes[1].set_title("Segment 1",fontsize=18)
        #axes[1].legend()
        axes[1].set_xlabel("Frequency (Hz)",fontsize=17)
        axes[1].set_ylabel("Ionic conductivity (S/cm)",fontsize=17)
        axes[1].set_xscale("log")
        axes[1].set_yscale("log")
        #axes[1].set_ylim(1e-8,1e-2)
        axes[1].grid(color='grey',ls=':',lw=1)
    elif nseg>1:
        for i in range(nseg):
            idx_low = len(x1)//nseg*i
            idx_high = len(x1)//nseg*(i+1)+1
            axes[i,0].plot(x1[idx_low:idx_high],y1[idx_low:idx_high],'ko')
            axes[i,0].set_title("Segment {}".format(i+1),fontsize=18)
            axes[i,0].set_xlabel("1000/T $\mathregular{(K^{-1})}$",fontsize=17)
            axes[i,0].set_ylabel("Ionic conductivity (S/cm)",fontsize=17)
            axes[i,0].set_yscale("log")
            axes[i,0].grid(color='grey',ls=':',lw=1)

            for j in range(idx_low,idx_high):
                axes[i,1].plot(x2[j],y2[j],c=colors[j])
                axes[i,1].plot(x3[j],y1[j],marker='x',c='k')
            axes[i,1].set_title("Segment {}".format(i+1),fontsize=18)
                #axes[i,1].legend()
            axes[i,1].set_xlabel("Frequency (Hz)",fontsize=17)
            axes[i,1].set_ylabel("Ionic conductivity (S/cm)",fontsize=17)
            axes[i,1].set_xscale("log")
            axes[i,1].set_yscale("log")
            axes[i,1].grid(color='grey',ls=':',lw=1)

        
    fig.tight_layout()

plot_conductivities(fig_x1,fig_y1,fig_x2,fig_y2,fig_x3,num_segment)
plt.savefig(f"{output_name}.png",dpi=300)
plt.show()
```
