```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import glob
import seaborn as sns
import numpy as np
#sns.set()

file_name_list = []
for file_name in glob.glob('Chosen to use/*PCL*.csv'):
    file_name_list.append(file_name)
    
file_name_list.sort(reverse = True)
print(file_name_list)
len(file_name_list)
```

```python
# set default font style and size globally in matplotlib
plt.rcParams['font.family'] = 'Times New Roman'
plt.rcParams['font.size'] = 18
plt.rcParams['figure.figsize'] = 6, 4.8

# apply color maps
n = 7
colors = plt.cm.Blues(np.linspace(1,0.2,n)) # Blues

"""
df = pd.read_csv('0 MgTFSI2 PEO.csv')
temperature = df.iloc[7650:len(df)-1000, 1].astype('float').values.tolist()
heatflow = df.iloc[7650:len(df)-1000, 2].astype('float')
plt.plot(temperature, heatflow)
"""

# plot data
for i in range(len(file_name_list)):
    df = pd.read_csv(file_name_list[i])
    temperature = df.iloc[7650:len(df)-1000, 1].astype('float').values.tolist()
    heatflow = df.iloc[7650:len(df)-1000, 2].astype('float')
    
    # Shift the data
#    heatflow = heatflow/(heatflow[len(heatflow)-1]-heatflow[200])
    heatflow = -1*heatflow - (i - len(file_name_list))
    if i==2:
        heatflow = heatflow - 1
    elif i==1:
        heatflow = heatflow + 0.5
    heatflow = heatflow.values.tolist()

    filename = file_name_list[i]
    
    plt.plot(temperature, heatflow, color=colors[i], label=filename[14:-4])
plt.xlabel('$\mathregular{Temperature\; (^{o}C)}$', fontsize=20)
plt.ylabel('Heat Flow', fontsize=20)
plt.ylim(1, 9)
#plt.yticks(np.arange(1,10,2))
plt.yticks([])
plt.xlim(-70, 100)

ax = plt.gca()
#ax.axes.xaxis.set_ticklabels([])
ax.axes.yaxis.set_ticklabels([])

#plt.legend(loc='lower right',bbox_to_anchor=(0.5, 0.5, 1, 1), prop={'size': 10})#,bbox_to_anchor=(0.5, 0.5, 1, 1),prop={'size': 10},ncol=1)
#plt.grid(b=None, which='major', axis='both', color='grey', linestyle='--', linewidth=0.5)
plt.tight_layout()
plt.savefig(file_name_list[0][14:-4]+' all samples.png', dpi=300)

print(temperature[200], temperature[len(temperature)-1])
```

```python
from scipy.signal import savgol_filter

# set default font style and size globally in matplotlib
plt.rcParams['font.family'] = 'Times New Roman'
plt.rcParams['font.size'] = 18
plt.rcParams['figure.figsize'] = 6, 4.8

# apply color maps
n = 7
colors = plt.cm.Blues(np.linspace(1,0.2,n)) # Blues

"""
df = pd.read_csv('0 MgTFSI2 PEO.csv')
temperature = df.iloc[7650:len(df)-1000, 1].astype('float').values.tolist()
heatflow = df.iloc[7650:len(df)-1000, 2].astype('float')
plt.plot(temperature, heatflow)
"""

# plot data
for i in range(len(file_name_list)):
    df = pd.read_csv(file_name_list[i])
    temperature = df.iloc[7650:len(df)-1000, 1].astype('float').values.tolist()
    heatflow = df.iloc[7650:len(df)-1000, 2].astype('float')
    
    heatflow_raw = -np.array(heatflow)
    
    # Shift the data
#    heatflow = heatflow/(heatflow[len(heatflow)-1]-heatflow[200])
    heatflow = -1*heatflow - (i - len(file_name_list))
    if i==2:
        heatflow = heatflow - 1
    elif i==1:
        heatflow = heatflow + 0.5
    heatflow = heatflow.values.tolist()

    heatflow = np.array(heatflow)
    deriv_heatflow = 700*savgol_filter(heatflow_raw, window_length=201, polyorder=2, deriv=1) # derivative and smooth
    
    filename = file_name_list[i]
    
    deriv_heatflow = deriv_heatflow - (i - len(file_name_list))
    
    plt.plot(temperature, deriv_heatflow, color=colors[i], label=filename[14:-4])
plt.xlabel('$\mathregular{Temperature\; (^{o}C)}$', fontsize=20)
plt.ylabel('Diff. Heat Flow', fontsize=20)
plt.ylim(-1, 9)
#plt.yticks(np.arange(1,10,2))
plt.yticks([])
plt.xlim(-70, 100)

ax = plt.gca()

ax.axes.yaxis.set_ticklabels([])

#plt.legend(loc='lower right',bbox_to_anchor=(0.5, 0.5, 1, 1), prop={'size': 10})#,bbox_to_anchor=(0.5, 0.5, 1, 1),prop={'size': 10},ncol=1)
#plt.grid(b=None, which='major', axis='both', color='grey', linestyle='--', linewidth=0.5)
plt.tight_layout()
plt.savefig(file_name_list[0][14:-4]+'deriv all samples.png', dpi=300)

```
