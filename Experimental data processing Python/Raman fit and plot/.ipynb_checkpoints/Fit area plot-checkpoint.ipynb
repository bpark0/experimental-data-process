```python
import matplotlib.pyplot as plt
%matplotlib inline
import pandas as pd
import glob
import os
import numpy as np

num_peaks = 4 # number of peaks you want to see (e.g. 4 for 4 peaks)
#what = 1 # index of what_to_plot, change it (from 0 to 2)
#what_to_plot = ['area', 'position', 'chi_square']
salt_and_polymer = 'MgTFSI PCL-PTMC '

### True when saving ###
save_results = False

### which computer are you using? (imac? 1st macbook? 2nd) ###
#user_directory = '/Users/bumjunpark/'
user_directory = '/Users/bagbeomjun/'

file_name_list = []
directory_list = []
full_list = []
for folderName, subfolders, filenames in os.walk(user_directory+'Library/Mobile Documents/com~apple~CloudDocs/Research/Data/Raman/10042019 (from 0524)/Mg-PCL-PTMC/Chosen data'): # returns three values during for loop
    #print('The folder is ' + folderName + '\n')
    #print('The subfolders in ' + folderName + ' are: ' + str(subfolders) + '\n')
    #print('The filenames in ' + folderName + ' are: ' + str(filenames) + '\n')
    #print('\n\n')
    
    # it is easy to make another for loop to do a specific job
    for subfolder in subfolders:
        #print(subfolder)
        if not subfolder.endswith('checkpoints'):
            directory_list.append(subfolder)
   
    for file in filenames:
        if file.endswith('infos.csv'):
            file_name_list.append(file)
            #shutil.copy(os.path.join(folderName, file), os.join(folderName, file) + '.backup')
            #print('copy this file: ' + os.path.join(folderName, file) + '\ninto: ' + os.path.join(folderName, file) + '.backup')
            full_list.append(os.path.join(folderName, file))
directory_list.sort(reverse=True)
file_name_list.sort(reverse=True)
full_list.sort(reverse=True)


############## plot ##############
# set default font style and size globally in matplotlib
plt.rcParams['font.family'] = 'Times New Roman'
plt.rcParams['font.size'] = 16
plt.rcParams['figure.figsize'] = 6.4, 4.8

file_peaks_3 = []
file_peaks_4 = []
file_peaks_5 = []
file_peaks_6 = []
for directory in full_list:
    folder = directory.split('/')[-2]
    file = directory.split('/')[-1]
    if not folder.startswith('practice'):
        if folder.endswith('3 peaks'):
            file_peaks_3.append(os.path.join(folder, file))
        elif folder.endswith('4 peaks'):
            file_peaks_4.append(os.path.join(folder, file))
        elif folder.endswith('5 peaks'):
            file_peaks_5.append(os.path.join(folder, file))
        elif folder.endswith('6 peaks'):
            file_peaks_6.append(os.path.join(folder, file))

file_peaks_3.sort()
file_peaks_4.sort()
file_peaks_5.sort()
file_peaks_6.sort()

print(file_peaks_3)

file_list = []
file_list.append(file_peaks_3)
file_list.append(file_peaks_4)
file_list.append(file_peaks_5)
file_list.append(file_peaks_6)

peak_area = []
peak_position = []
chi_square = []
salt_conc = [8, 12, 16, 20, 28, 36]

for i in range(len(file_list[num_peaks-3])): # change the list to plot different peak numbers
    data = pd.read_csv(file_list[num_peaks-3][i]) # change the list to plot different peak numbers
    area = list(data.iloc[:, 0])
    peak_loc = list(data.iloc[:, 1])
    chi_sq = list(data.iloc[:, 2])
    if area[1] > 100:
        area_temp = area
        area = peak_loc
        peak_loc = area_temp
    
    peak_area.append(area)
    peak_position.append(peak_loc)
    chi_square.append(chi_sq)

peak_area = np.array(peak_area).transpose()
peak_position = np.array(peak_position).transpose()
chi_square = np.array(chi_square).transpose()

print('number of peaks:', len(peak_area))
print('number of concentrations:', len(peak_area[0]))

markers = ['o', 's', '^', 'v', '<', '>']
colors = ['r', 'b', 'g', 'y', 'm']
labels = ['Peak 1', 'Peak 2', 'Peak 3', 'Peak 4', 'Peak 5', 'Peak 6']


# set default font style and size globally in matplotlib
plt.rcParams['font.family'] = 'Times New Roman'
plt.rcParams['font.size'] = 15
plt.rcParams['figure.figsize'] = 6.4, 4.8

for i in range(len(peak_area)):
    plt.plot(salt_conc[:2], peak_area[i][:2], linestyle='None', marker=markers[i], color=colors[i], label=labels[i])
    plt.plot(salt_conc[2:4], peak_area[i][2:4], linestyle='None', marker=markers[i], markerfacecolor='None', color=colors[i])
    plt.plot(salt_conc[4:len(peak_area[0])], peak_area[i][4:], linestyle='None', marker=markers[i], color=colors[i])

plt.xlim(5,40)
plt.ylim(-0.05,1.05)
plt.xlabel('Salt concentration (wt.%)', fontsize=18)
plt.ylabel('Area ratio', fontsize=18)

plt.legend(loc='best',fontsize=15)
plt.locator_params(axis='x', nbins=13)
plt.grid(b=None, which='major', axis='both', color='grey', linestyle='--', linewidth=0.5)
plt.tight_layout()
if save_results == True:
    plt.savefig(salt_and_polymer + 'Salt concentration ' + str(num_peaks) + ' peaks.png', dpi=300)
plt.show()
```
