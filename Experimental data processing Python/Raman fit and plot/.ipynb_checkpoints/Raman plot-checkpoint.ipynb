```python
import pandas as pd
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
import glob
import numpy as np

# input the file name
file_name_list = []
for file_name in glob.glob('*.csv'):
    file_name_list.append(file_name)
    
file_name_list.sort(reverse = True)
print(file_name_list)

```

```python
# define the range of data in the .csv file as the number of rows
start_of_dataset = 19
end_of_dataset = 3803

# Functions needed

def data_process(file_name, shift = 0):
    xaxis = 'Wavenumber (cm$^-1$)'
    yaxis = file_name
    data = pd.read_csv(file_name, names=[xaxis,yaxis[:-4]])
    data_trim_original = data[start_of_dataset:end_of_dataset]
    data_trim_original = data_trim_original.astype(float)
    data_trim = data_trim_original.set_index(xaxis)
    data_norm = data_trim/(data_trim[600:800].max())
    data_norm = data_norm + shift
    return data_norm

def data_process_normbymax(file_name, shift = 0):
    xaxis = 'Wavenumber (cm$^-1$)'
    yaxis = file_name
    data = pd.read_csv(file_name, names=[xaxis,yaxis[:-4]])
    data_trim_original = data[start_of_dataset:end_of_dataset]
    data_trim_original = data_trim_original.astype(float)
    data_trim = data_trim_original.set_index(xaxis)
    data_norm = data_trim/(data_trim.max())
    data_norm = data_norm + shift
    return data_norm

# find peak ratio (functional group vs. TFSI- shift)
def find_peak_ratio(file_name, lower_limit, upper_limit):
    xaxis = 'Wavenumber (cm^-1)'
    yaxis = file_name
    data = pd.read_csv(file_name, names=[xaxis,yaxis[:-4]])
    data_trim_original = data[start_of_dataset:end_of_dataset]
    data_trim_original = data_trim_original.astype(float)
    data_trim = data_trim_original.set_index(xaxis)
    peak_ratio = (data_trim[lower_limit:upper_limit].max())/(data_trim[600:800].max())
    peak_index = data_trim[600:800].idxmax()
    return peak_ratio

def find_peak_index(file_name, lower_limit, upper_limit):
    xaxis = 'Wavenumber (cm^-1)'
    yaxis = file_name
    data = pd.read_csv(file_name, names=[xaxis,yaxis[:-4]])
    data_trim_original = data[start_of_dataset:end_of_dataset]
    data_trim_original = data_trim_original.astype(float)
    data_trim = data_trim_original.set_index(xaxis)
    peak_index = data_trim[600:800].idxmax()
    return peak_index
```

```python
# set default font style and size globally in matplotlib
plt.rcParams['font.family'] = 'Times New Roman'
plt.rcParams['font.size'] = 15
plt.rcParams['figure.figsize'] = 6.4, 4.8

# apply color maps
n = 6
colors = plt.cm.Blues(np.linspace(1,0.4,n)) # Blues

# plot data
for i in range(len(file_name_list)-1):
    df = data_process(file_name_list[i]) - 0.5*(i+1) + 0.5*(len(file_name_list)-1)
    filename = file_name_list[i]
    plt.plot(df, color=colors[i], label=filename[:-4])
    
plt.legend(loc='upper left',bbox_to_anchor=(1, 0, 1, 1),prop={'size': 12})
plt.ylabel('Intensity (a.u.)', fontsize=18)
plt.xlabel('Wavenumber (cm$^{-1}$)', fontsize=18)
# plt.xlim(1910,9) # full scale
plt.xlim(800,700) # peak scale
plt.ylim(-0.05, 0.5*(len(file_name_list))+0.05)
plt.locator_params(axis='y', nbins=8)
plt.locator_params(axis='x', nbins=5)
plt.grid(b=None, which='major', axis='both', color='grey', linestyle='--', linewidth=0.5)
plt.savefig('MgTFSI2 PCL-PTMC all peak.png', dpi=300) # peak scale
plt.show()
```
